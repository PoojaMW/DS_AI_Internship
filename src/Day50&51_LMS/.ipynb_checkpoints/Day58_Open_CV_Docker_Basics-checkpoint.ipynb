Work Summary

Today, I explored two foundational pillars of modern AI engineering: Computer Vision and Containerization. I learned how to use OpenCV to interpret images as numerical matrices (NumPy arrays) and performed essential preprocessing tasks like grayscale conversion, resizing, and blurring. In the second half of the day, I tackled environment standardization using Docker. I learned to wrap my Python applications into lightweight, portable Containers, ensuring that my code runs identically regardless of the host operating system. I successfully wrote a Dockerfile, built a custom image, and executed my image-processing script within a containerized environment.

Hours worked 4.5

Learnings / Outcomes
Images as Data: Realized that in the eyes of a computer, an image is just a grid of numbers. I learned that color images are 3D arrays (Blue, Green, Red channels), while grayscale images are 2D, which significantly reduces computational load for ML models.
The Preprocessing Pipeline: Mastered basic OpenCV functions (cvtColor, resize, GaussianBlur) to clean and standardize visual data before it reaches a model.
Containerization vs. Virtualization: Understood that Docker containers are superior to Virtual Machines because they share the host OS kernel, making them much faster and more resource-efficient.
Infrastructure as Code: Practiced writing a Dockerfile, which acts as a "recipe" for my environment. By defining my base image (python:3.9) and dependencies, I can now recreate my exact workspace on any machine in seconds.

Blockers / Risks
The "GUI" Container Headache: I discovered that running cv2.imshow() inside a standard Docker container often fails because containers don't have a built-in display (X11 forwarding). I learned that for containerized vision apps, it's often better to save the output to a file (cv2.imwrite) rather than trying to pop up a window.
Color Channel Confusion: A common "gotcha" I encountered is that OpenCV reads images in BGR (Blue-Green-Red) format by default, while most other libraries (like Matplotlib) expect RGB. I have to be mindful of this swap during visualization.
Docker Context & Image Size: My first Docker image was unnecessarily large. I realized that the COPY . . command includes everything in the folder, I need to be careful about what files are in my build context to keep images lean.

Skills Used

Computer Vision (OpenCV), Image Matrix Manipulation, Environment Standardization (Docker), Containerization, Dockerfile Scripting, Image Preprocessing, Python Backend Development.
