Work Summary

Today, I explored the foundational architecture of Neural Networks, starting with the biological inspiration of the human brain and its transition into the mathematical Artificial Neuron. I spent the day deconstructing the Perceptron—the simplest form of a neural network—and implemented its logic manually in Python. I learned how inputs are modified by Weights (importance) and Biases (thresholds) before being passed through Activation Functions like Sigmoid and ReLU to introduce non-linearity. I concluded the day by moving from a single neuron to a Multi-Layer Network structure, performing manual forward propagation to see how data flows from an input layer through hidden layers to produce a final prediction.

Hours worked 4.5

Learnings / Outcomes
Biological Mapping: Understood that the Artificial Neuron is a mathematical abstraction where Dendrites become Inputs, the Cell Body becomes a Summation function, and the Axon becomes the Output.
The Perceptron Logic: Successfully modeled logic gates (AND/OR) using a single perceptron. This taught me that a neuron is essentially a Linear Decision Boundary—it "slices" space to separate classes.
Non-Linearity via Activation: Realized that without functions like ReLU or Sigmoid, a neural network is just a giant linear equation, no matter how many layers it has. Activation functions are what allow the network to learn complex, "curvy" patterns in data.

Blockers / Risks
The Linearity Trap: I initially struggled to understand why we need activation functions. After running an experiment without them, I saw that the network couldn't solve a simple non-linear problem (like XOR). This proved that Non-linearity is the "secret sauce" of deep learning.
Manual Calculation Errors: While performing forward propagation manually, I found it easy to lose track of the bias term or forget to apply the activation function at the very end. This highlighted why we eventually use libraries like NumPy to handle these matrix operations.
Weight Initialization: I noticed that if I set my weights to zero, the network wouldn't learn effectively. I learned that "Initial Weights" significantly impact how quickly (or if) a model converges to a solution.

Skills Used

Neural Network Foundations, Perceptron Implementation, Mathematical Modeling (Weighted Sums), Non-Linear Activations (Sigmoid/ReLU), Forward Propagation, NumPy Matrix Operations, Logic Gate Modeling.
