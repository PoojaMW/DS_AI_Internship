```python
## Work Summary

Today, I moved beyond the limitations of a single train-test split to master **Cross-Validation (CV)**. I learned that relying on one random split can lead to unstable and misleading performance estimates. By implementing **K-Fold Cross-Validation**, I practiced evaluating models multiple times on different subsets to ensure every data point serves as both training and testing data. I also explored specialized techniques like **Stratified K-Fold** for imbalanced datasets and **Leave-One-Out (LOOCV)** for exhaustive testing. A key focus of the day was identifying and preventing **Data Leakage**, ensuring that preprocessing steps like scaling do not "peek" at the test data before evaluation.

**Hours worked**
4.5

**Reference Links**
  https://github.com/PoojaMW/DS_AI_Internship/blob/main/src/Day45_Cross_Validation/Cross_Validation.ipynb
  * [Scikit-Learn: Cross-Validation Guide](https://scikit-learn.org/stable/modules/cross_validation.html)

-----

## Learnings / Outcomes

  * **The Variance Problem:** Observed how a model's accuracy can swing significantly just by changing the `random_state` in a standard split. Cross-validation provides a stable **Mean Score** and **Variance** to quantify this uncertainty.
  * **K-Fold Mechanism:** Implemented the process of dividing data into $K$ folds, training on $K-1$ and testing on the remainder. This ensures the evaluation is representative of the entire dataset.
  * **Stratification Strategy:** Learned that for classification, **Stratified K-Fold** is mandatory to maintain the original class ratio in every fold, preventing "lucky" folds where a minority class might be missing entirely.
  * **LOOCV (Leave-One-Out):** Understood that while LOOCV is computationally expensive (testing on one sample at a time), it provides a near-unbiased estimate for very small datasets.
  * **Data Leakage Awareness:** Realized that scaling or transforming the entire dataset before splitting "leaks" information from the test set into the training process, leading to over-optimistic (and false) results.

-----

## Blockers / Risks

  * **Computational Cost:** During the LOOCV session, I noticed a significant jump in processing time. I learned that for large datasets, $K=5$ or $K=10$ is the industry standard balance between speed and statistical rigor.
  * **The Preprocessing Trap:** I initially scaled my data *before* running `cross_val_score`. I had to correct this by understanding that scaling parameters ($mu$, $sigma$) must only be derived from the training folds to avoid leakage.
  * **Interpretability:** I initially struggled with how to use the CV results. I now understand they are for **model selection** and **validation**, not for the final production model itself.

-----

## Skills Used

K-Fold Cross-Validation, Stratified K-Fold, Leave-One-Out (LOOCV), Data Leakage Prevention, Model Performance Estimation, Scikit-learn (`model_selection` module).
```
