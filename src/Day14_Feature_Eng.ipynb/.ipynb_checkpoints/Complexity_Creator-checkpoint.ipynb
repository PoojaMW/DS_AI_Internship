```python


import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score

# Example dataset (non-linear relationship)
# House size vs price (curved pattern)
X = np.array([1,2,3,4,5,6,7,8]).reshape(-1,1)
y = np.array([50,60,65,80,110,150,210,280])  # non-linear growth

# -----------------------------
# 1️⃣ Linear Regression (Original Feature)
# -----------------------------
model_linear = LinearRegression()
model_linear.fit(X, y)
y_pred_linear = model_linear.predict(X)
r2_linear = r2_score(y, y_pred_linear)

# -----------------------------
# 2️⃣ Polynomial Features (degree = 2)
# -----------------------------
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model_poly = LinearRegression()
model_poly.fit(X_poly, y)
y_pred_poly = model_poly.predict(X_poly)
r2_poly = r2_score(y, y_pred_poly)

# Results
print("R² Score (Linear):", r2_linear)
print("R² Score (Polynomial):", r2_poly)
```
