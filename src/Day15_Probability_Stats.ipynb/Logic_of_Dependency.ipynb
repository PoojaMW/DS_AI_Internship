```python
# -------------------------------
# Independent Events
# Coin flip (Heads) AND Die roll (6)
# -------------------------------

P_heads = 1 / 2      # Probability of Heads
P_six = 1 / 6        # Probability of rolling a 6

P_independent = P_heads * P_six
print("Probability of Heads AND rolling a 6:", P_independent)


# -------------------------------
# Dependent Events
# Two red marbles without replacement
# -------------------------------

red_marbles = 5
blue_marbles = 5
total_marbles = red_marbles + blue_marbles

# First marble is red
P_first_red = red_marbles / total_marbles

# Second marble is red (after one red is removed)
P_second_red = (red_marbles - 1) / (total_marbles - 1)

P_dependent = P_first_red * P_second_red
print("Probability of drawing two red marbles:", P_dependent)
```
